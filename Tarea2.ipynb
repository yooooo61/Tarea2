```python
from ColabTurtle.Turtle import * #Importamos todas las funciones de la librería
initializeTurtle() #Iniciamos turtle
#Definimos nuestras constantes
l = 100
a = 60
color('purple') #Cambiamos el color de la tortuga
right(90) #Giramos la tortuga a la derecha 90 grados
forward(l) #Avanzamos l espacios

##Repetimos estos 3 pasos cambiando el color para generar el hexágono
color('blue') #Volvemos a cambiar de color de la tortuga
left(a) #Hacemos que gire a la izquierda 60 grados
forward(l) #Hacemos que avance l espacios

color('green')
left(a)
forward(l)

color('yellow')
left(a)
forward(l)

color('orange')
left(a)
forward(l)

color('red')
left(a)
forward(l)
```

```python
from ColabTurtle.Turtle import * #Importamos todas las funciones de la librería
#Definimos nuestras constantes
l = 100
a = 60
initializeTurtle(initial_speed=8) #Iniciamos turtle a una velocidad de 8
hideturtle() #Ocultamos la tortuga
penup() #Evitamos que se dibujen lineas
#Ponemos el centro del hexágono en el centro de la pantalla
forward(l/2) 
left(90)
forward(l/1.15)
right(90)

pendown() #Hacemos que se dibujen las lineas

colors = ['purple','blue','green','yellow','orange','red'] #Generamos una lista con el nombre de nuestros colores

#Utilizamos un ciclo FOR para generar nuestro hexágono haciendo que gire "a" grados y que avance "l" espacios
for i in range(len(colors)):
  color(colors[i])
  right(a)
  forward(l)
```

```python
#Importamos las librerías
import pandas as pd
import numpy as np
#Creamos un Dataframe con el archivo que cargamos, le indicamos cual es el tipo de separación
df = pd.read_csv('autos.csv',sep=';')
#Pedimos la información del dataframe
df.info()
```

```python
#Al ver que son pocas las filas con datos nulos, creamos un dataframe eliminando esas filas
df_1=df.dropna()
#Pedimos la información del nuevo DF
df_1.info()
```

```python
#Generamos las nuevas columnas para nuestro nuevo DF con las condiciones que nos piden
#Yo identifiqué "Millas por galón" y "Cilindros" como mis variables
df_1["Var1"] = np.log(df_1['Millas_Por_Galon']/df_1['Cilindros'])

df_1["Var2"] = np.sqrt(df_1['Millas_Por_Galon'])*np.exp(df_1['Cilindros'])/200

df_1['Var3'] = np.where(df_1['Millas_Por_Galon']>df_1['Cilindros'],5,3)

df_1['Var4'] = 1/np.log(df_1['Millas_Por_Galon']/df_1['Cilindros'])*100

df_1['Var5'] = df_1['Cilindros']**2/df_1['Millas_Por_Galon']

df_1
```

```python
#Reiniciamos el nombre de las filas
df_1 = df_1.reset_index()
#Eliminamos la columna index generada en el paso anterior
df_1 = df_1.drop(["index"],axis="columns" )
df_1
```

```python
import statistics as stats
#Ponemos como Default el el valor "Mediano" para esta columna
df_1["Cualitativas1"] = "Mediano"
#Ahora le cambiamos con los valores "Alto" y "Bajo" como se requirió en el problema
df_1["Cualitativas1"] = np.where((df_1['Millas_Por_Galon'] > (np.mean(df_1["Millas_Por_Galon"])+stats.stdev(df_1['Millas_Por_Galon']))) | (df_1['Millas_Por_Galon'] < (np.mean(df_1["Millas_Por_Galon"])-stats.stdev(df_1['Millas_Por_Galon']))) ,"Alto","Bajo")

df_1
```

```python
import matplotlib.pyplot as plt
#Generamos un nuevo DF solo con las columnas que vamos a usar

df_bigote = df_1.loc[:, ['Millas_Por_Galon', 'Cilindros']] 
```

```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(7, 4))
#Generamos la gráfica tomando en cuenta los dos parámetros que vamos a utilizar
sns.boxplot(x = 'Cilindros', y = 'Millas_Por_Galon', data = df_bigote, ax = ax)
#Le damos el nombre a la gráfica
ax.set_title('Millas por galon por cilindraje');
```

```python
#Agrupamos las filas por cilindraje
df_barras = df_bigote.sort_values('Cilindros')
#Sacamos el promedio de cada grupo
df_barras_mean = df_barras.groupby("Cilindros",as_index=False).mean()
df_barras_mean
```

```python
#Creamos una gráfica del promedio de millas por galón por el cilindraje
plt.figure()
df_barras_mean.plot(x="Cilindros", y="Millas_Por_Galon",kind='bar')
#Creamos una gráfica de barras con una curva
plt.figure()
sns.distplot(df_barras['Millas_Por_Galon'],bins=30)
plt.show()
```

```python
#Creamos un DF quitando las variables con valor que no sea numérico
df_hm = df_1.drop(['Nombre','Cualitativas1'],axis=1)
#Generamos el mapa de calor relacionando todas nuestras columnas y le ponemos el color deseado
sns.heatmap(df_hm.corr(),cmap="cool")
```
